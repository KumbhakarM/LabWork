```
# Install the required libraries
#!pip install earthengine-api geemap

# Import the required libraries
import ee
import geemap

# Initialize the Earth Engine API (after authentication)
ee.Initialize(project='ee-mrityunjoy')

# Define the Area of Interest (AOI), replace with your specific coordinates
aoi = ee.FeatureCollection("projects/ee-mrityunjoy/assets/Samudratapu_2014")  # Adjust to your area coordinates

# Load a Landsat 8 image collection (Collection 2, Level 2) and filter by date and location
collection = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
                .filterDate('2023-01-01', '2023-12-31') \
                .filterBounds(aoi) \
                .median()  # Take the median image for the time period

# Select the Green and SWIR bands (Landsat 8 Collection 2 Level 2 uses SR_B3 for Green and SR_B6 for SWIR1)
green_band = collection.select('SR_B3')
swir_band = collection.select('SR_B6')

# Calculate NDSI
ndsi = green_band.subtract(swir_band).divide(green_band.add(swir_band)).rename('NDSI')

# Define a snow threshold, commonly set at 0.4
snow_threshold = 0.4
snow_mask = ndsi.gt(snow_threshold)  # Creates a binary mask where snow areas have a value of 1

# Apply the snow mask to the NDSI image
snow_area = ndsi.updateMask(snow_mask)

# Create a map using geemap
Map = geemap.Map()

# Create a map using geemap
#Map = geemap.Map(center=[77.37, 32.78], zoom=8)
Map.centerObject(aoi, zoom=8)

# Add the original NDSI layer for visualization
Map.addLayer(ndsi, {'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDSI')

# Add the masked snow area layer
Map.addLayer(snow_area, {'palette': ['blue']}, 'Snow Masked Area')

# Display the map
Map
```

```
# Install the required libraries
#!pip install earthengine-api geemap

# Import the required libraries
import ee
import geemap

# Initialize the Earth Engine API (after authentication)
ee.Initialize(project='ee-mrityunjoy')

# Load your study area as a FeatureCollection
aoi = ee.FeatureCollection("projects/ee-mrityunjoy/assets/Samudratapu_2014")  # Replace with your actual asset path

# Load a Landsat 8 image collection (Collection 2, Level 2) and filter by date and location
collection = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2') \
                .filterDate('2023-01-01', '2023-12-31') \
                .filterBounds(aoi) \
                .median()  # Take the median image for the time period

# Select the Green and NIR bands (Landsat 8 Collection 2 Level 2 uses SR_B3 for Green and SR_B5 for NIR)
green_band = collection.select('SR_B3')
nir_band = collection.select('SR_B5')

# Calculate NDWI
ndwi = green_band.subtract(nir_band).divide(green_band.add(nir_band)).rename('NDWI')

# Define a water threshold, commonly set at 0.3
water_threshold = 0.1
water_mask = ndwi.gt(water_threshold)  # Creates a binary mask where water bodies have a value of 1

# Apply the water mask to the NDWI image
water_bodies = ndwi.updateMask(water_mask)

# Create a map using geemap
Map = geemap.Map()

# Center the map on your AOI
Map.centerObject(aoi, zoom=8)

# Add the original NDWI layer for visualization
Map.addLayer(ndwi, {'min': -1, 'max': 1, 'palette': ['blue', 'white', 'green']}, 'NDWI')

# Add the masked water bodies layer
Map.addLayer(water_bodies, {'palette': ['blue']}, 'Water Bodies')

# Display the map
Map
```
